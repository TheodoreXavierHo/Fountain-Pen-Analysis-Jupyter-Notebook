```python
# Function to check duplicates in a DataFrame
def check_duplicates(df):
    # Iterate through all columns of the DataFrame
    for col in df.columns:
        # Check if the column contains lists
        if isinstance(df[col].iloc[0], list):
            # Convert the column to strings
            df[col] = df[col].astype(str)

    # Check for duplicates in the DataFrame
    duplicates = df[df.duplicated()]

    # Print the result
    if duplicates.empty:
        # If there are no duplicates, print a message
        print("No duplicates found")
    else:
        # If there are duplicates, print the duplicate rows
        print("Duplicate Rows:")
        print(duplicates)

# Check for duplicates in the DataFrame
check_duplicates(df_allfountainpen)
```

## 2.5: Get Specifications and Content from Product Links
### Extracting Detailed Product Information
In this step, we extract more detailed information such as specifications and content descriptions from each product's individual link. A ThreadPoolExecutor is used for efficient concurrent processing. This rich data will help us understand more nuanced details about each pen, which could be influential in their popularity.
- Define a function to extract specifications and content from product links.
- Use a ThreadPoolExecutor to process multiple product links concurrently.
- Create new columns 'specifications' and 'content' in the DataFrame.

```python
# Function to extract specifications and content from a product link
def get_data(product_link):
    response = requests.get(product_link)
    if response.status_code == 200:
        html = response.content
        soup = BeautifulSoup(html, 'html.parser')

        # Get specifications
        specs_list_element = soup.find('dl', class_='specs-list')
        specs_dict = {}
        dt_elements = specs_list_element.find_all('dt', class_='specs-list__item')
        for dt in dt_elements:
            key = dt.get_text(strip=True)
            value = dt.find_next('dd').get_text(strip=True)
            specs_dict[key] = value

        # Get content
        content_element = soup.find('div', class_='product-description js-productDesc')
        content = content_element.get_text(strip=True) if content_element else ''

        return {'specifications': specs_dict, 'content': content}
    else:
        return {'specifications': {}, 'content': ''}

# Use a thread pool to process product links concurrently
with ThreadPoolExecutor(max_workers=16) as executor:
    results = list(tqdm(executor.map(get_data, df_allfountainpen['url']), total=len(df_allfountainpen)))

# Create new columns 'specifications' and 'content' in the DataFrame
df_allfountainpen[['specifications', 'content']] = pd.DataFrame(results, index=df_allfountainpen.index)
```

## 2.6: Fetch API Links for Reviews
### Retrieving API Links for Product Reviews
Here, we construct API links for fetching product reviews. This step is essential for collecting user-generated content, which offers valuable insights into customer preferences and satisfaction.
- Define a function to fetch API links for reviews.
- Use a ThreadPoolExecutor to process API link generation concurrently.

```python
# Function to fetch API links for reviews
def fetch_api_link(row_tuple):
    product_name_with_plus = row_tuple.title.replace(' ', '+')
    
    api_url = "https://stamped.io/api/widget"
    params = {
        "productId": row_tuple.id,
        "productName": product_name_with_plus,
        "productSKU": row_tuple.handle,
        "page": "1",
        "apiKey": "pubkey-n640C2EL927aH2U8w318I967FLW2I7",
        "storeUrl": "thegouletpencompany.myshopify.com",
        "take": "5"
    }

    api_link = f"{api_url}?{('&'.join(f'{key}={value}' for key, value in params.items()))}"

    return api_link

# Function to apply the fetch_api_link function using ThreadPoolExecutor
def apply_api_requests(df):
    with ThreadPoolExecutor(max_workers=16) as executor:
        api_links = list(tqdm(executor.map(fetch_api_link, df.itertuples(index=False), chunksize=10), total=len(df), desc="Fetching API Links"))
    return api_links

# Create a new 'api' column with API links using ThreadPoolExecutor
df_fountain_pen_review = pd.DataFrame(apply_api_requests(df_allfountainpen), columns=['api'])
```

## 2.7: Fetch Review Data from API Links
### Gathering Review Data
Utilizing the previously constructed API links, this section is focused on retrieving review data such as rating values and review counts. This information is key to understanding consumer reception and popularity of the products.
- Define a function to get review data from API links.
- Use a ThreadPoolExecutor to process review data retrieval concurrently.
- Create new columns 'rating_value' and 'review_count' in the DataFrame.

```python
# Function to get review data from API links
def get_review_data(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            rating_value = soup.find('meta', itemprop='ratingValue')['content']
            review_count = soup.find('meta', itemprop='reviewCount')['content']
            return {'rating_value': rating_value, 'review_count': review_count}
        else:
            return {'rating_value': None, 'review_count': None}
    except Exception as e:
        return {'rating_value': None, 'review_count': None}

# Function to process URLs using ThreadPoolExecutor
def process_urls(urls):
    results = []
    with ThreadPoolExecutor(max_workers=16) as executor:
        futures = [executor.submit(get_review_data, url) for url in urls]
        for future in tqdm(futures, total=len(urls), desc="Processing URLs"):
            result = future.result()
            results.append(result)
    return results

# Process review API URLs using ThreadPoolExecutor
urls_to_process = df_fountain_pen_review['api'].str.strip().tolist()
results = process_urls(urls_to_process)

# Create new columns 'rating_value' and 'review_count' in the DataFrame
df_fountain_pen_review[['rating_value', 'review_count']] = pd.DataFrame(results, index=df_fountain_pen_review.index)
```

## 2.8: Fetch Customer Review Data from API Links
### Extracting In-Depth Customer Reviews
In addition to basic review data, we delve deeper into customer reviews, scraping detailed feedback from each product. This step is significant for a nuanced understanding of customer opinions and preferences.
- Define a function to get customer review data from API links.
- Use a ThreadPoolExecutor to process customer review data retrieval concurrently.
- Create a new DataFrame to store the review data.

```python
# Function to get customer review data from API links
def get_customer_review_data(url):
    all_reviews = []
    for pages in range(150):  # Hard Limit on the number of page reviews
        response = requests.get(url)
        try:
            if response.status_code == 200:
                html_content = response.text
                # Replace escaped quotes with regular quotes
                html_content = re.sub(r'\\"', '"', html_content)
                soup = BeautifulSoup(html_content, 'html.parser')

                reviews = soup.find_all(id=lambda value: value and value.startswith('stamped-review-'))
                    
                if not reviews:
                    if pages > 0:
                        return all_reviews
                    else:
                        return None

                for review in reviews:
                    try:
                        review_dict = {}
                        review_dict['author'] = review.find('strong', class_='author').text
                        review_dict['rating'] = review.get('data-rating')
                        review_dict['title'] = review.find('h3', class_='stamped-review-header-title').text
                        review_dict['content'] = review.find('p', class_='stamped-review-content-body').text
                        all_reviews.append(review_dict)
                    except Exception as e:
                        break
                
        except Exception as e:
            return None
        
        parsed_url = urlparse(url)
        query_params = parse_qs(parsed_url.query)
        page_number = int(query_params.get('page', ['1'])[0])
        updated_page_number = page_number + 1
        query_params['page'] = [str(updated_page_number)]
        parsed_url = parsed_url._replace(query=urlencode(query_params, doseq=True))
        url = urlunparse(parsed_url)               
    return all_reviews

# Function to process API URLs for customer reviews using ThreadPoolExecutor
def process_api_urls(urls):
    results = []
    with ThreadPoolExecutor(max_workers=16) as executor:
        futures = [executor.submit(get_customer_review_data, url) for url in urls]
        each_pen_review = []
        for future in tqdm(futures, total=len(urls), desc="Processing URLs"):
            result = future.result()
            each_pen_review.append(result)
        results.append(each_pen_review)
    return results

# Process customer review API URLs using ThreadPoolExecutor
all_reviews_data = process_api_urls(urls_to_process)

# Create DataFrame from the accumulated reviews
df_all_reviews_info = pd.DataFrame(all_reviews_data).T
df_all_reviews_info.columns = ['review_data']
```

### 2.16.1: Cleaning and Transforming Specification Data
In this section, we further clean the specification data by converting relevant fields to numeric types where applicable. This includes parsing numeric values from strings and handling missing or irregular data. We also consolidate similar columns to simplify the dataset.
- Cleaning and transforming specification data for fountain pens is the focus of this section.
- Numeric fields are converted to numeric types where applicable, involving parsing numeric values from strings and addressing missing or irregular data.
- Relevant columns related to Max Ink Capacity are combined and converted to numeric type (float).
- Additional columns such as 'Brand,' 'Condition,' and 'Type' are dropped from the DataFrame.
- A new 'Max Ink Capacity' column is created, containing the maximum value among specific columns, simplifying the dataset.

```python
# Define columns related to Max Ink Capacity to combine and convert to numeric type
columns_to_combine = [
    'max_ink_capacity_cartridge', 
    'max_ink_capacity_converter', 
    'max_ink_capacity_piston_vac', 
    'max_ink_capacity_eyedropper'
]

# Additional columns to drop
additional_columns_to_drop = ['brand', 'condition', 'type']

# Function to clean and convert values to numeric type (float)
def clean_and_convert(value):
    try:
        # Extract numeric parts from the value using regular expressions
        numeric_parts = re.findall(r'[\d.]+', str(value))
        # Join the numeric parts and convert to float
        return float(''.join(numeric_parts)) if numeric_parts else None
    except (ValueError, TypeError):
        return None

# Apply the conversion function to relevant columns
for col in columns_to_combine:
    if col in df_specifications.columns:
        df_specifications[col] = df_specifications[col].apply(clean_and_convert)
    else:
        print(f"Column '{col}' not found in df_specifications")

# Create a new column 'max_ink_capacity' with the maximum value among the specified columns
df_specifications['max_ink_capacity'] = df_specifications[columns_to_combine].max(axis=1, skipna=True)

# Combine lists of columns to drop
columns_to_drop = columns_to_combine + additional_columns_to_drop

# Drop the specified columns
df_specifications.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

### 3.1.2: Standardizing Text Formatting
This section focuses on enhancing the uniformity and readability of text data within the `all_fountain_pen_df` DataFrame. The main steps include cleaning up the review content and restructuring the review data format:
- **Clean up Review Data from Control Characters:**
  - Each review in the `review_data` column is processed to remove control characters like `\r` (carriage return) and `\n` (new line).
  - These characters are replaced with spaces to ensure consistent and clean text formatting.
- **Reformat and Combine Review Data:**
  - The `review_data` column, which contains lists of dictionaries, is reformatted.
  - Each review's 'title' and 'content' are combined into a single text, separated by a space, to create a more cohesive review text.
  - The 'author' field is removed from each review to maintain anonymity, focusing on the content and rating only.
  - The resulting structure for each review includes just the 'rating' and the combined 'content'.
These steps are crucial for preparing the dataset for further text analysis, ensuring that the review content is clean and uniformly formatted.


```python
# Step 1: Clean up Review Data from control characters

def clean_review_content(review_list):
    # Iterate through each review in the list
    for review in review_list:
        # Replace \r and \n with a space
        review['content'] = review['content'].replace('\\r', ' ').replace('\\n', ' ')
    return review_list

# Apply the cleaning function to the 'review_data' column
all_fountain_pen_df['review_data'] = all_fountain_pen_df['review_data'].apply(ast.literal_eval).apply(clean_review_content)
```

```python
# Step 2: Reformat the Review Data and combine the Title header with the Content body
def process_review_data(review_list):
    processed_reviews = []
    for review in review_list:
        # Combine 'title' and 'content' with a space, and remove 'author'
        combined_review = review.get('title', '') + ' ' + review.get('content', '')
        processed_review = {'rating': review.get('rating'), 'content': combined_review.strip()}
        processed_reviews.append(processed_review)
    return processed_reviews

# Apply the function to each row in the 'review_data' column
all_fountain_pen_df['review_data'] = all_fountain_pen_df['review_data'].apply(process_review_data)
```

### 3.3: Text Data Processing
Text data processing involves several steps to clean and prepare the text data from product reviews for further analysis and natural language processing tasks. These steps include removing stopwords, lemmatizing words, and cleaning specific phrases and punctuation.
- The operations are performed using the `progress_apply` method from `tqdm` to track the progress of the cleaning steps.

#### 3.3.1: Word Frequency Analysis
    - Create a function `get_word_frequency` to calculate the frequency of each word in the given column of a DataFrame.
    - Display the most common words in the 'content' column of both DataFrames.

```python
# Step 1: Word Frequency Analysis
def get_word_frequency(df, column):
    # Combine all text into one large string
    all_text = ' '.join(df[column].dropna())

    # Use regular expression to find words in the text
    words = re.findall(r'\w+', all_text.lower())

    # Use Counter to count the frequency of each word
    word_freq = Counter(words)

    # Return the dictionary containing word frequencies
    return word_freq
```

    - Define a function `clean_text_and_punctuation` to remove these words and punctuation marks from the text.
    - Apply this function to the 'content' column of both DataFrames to clean the text further.

```python
# Step 2: Cleaning Specific Phrases and Punctuation
def clean_text_combined(text):
    # Correct contractions (e.g., "wouldn ’ t" to "wouldn’t")
    contraction_pattern = re.compile(r"(?<=\w)\s’\s(?=\w)")
    text = re.sub(contraction_pattern, "’", text)

    # List of specific words and punctuation to remove
    words_to_remove = {
        'fountain', 'pen', 'another', 'ink', 'use', 'shop', 'ensure', 'here', 
        'get', 'would', 'even', 'bought', 'bit', 'got', 'one', 'due', 'comes', 'see', 
        'click', 'really', 'it', 'first', 'I', 'every', 'lot', '“', '”', '...', '  '
    }
    punctuation_to_replace = string.punctuation

    # Replace punctuation with space and remove specified words
    for punct in punctuation_to_replace:
        text = text.replace(punct, ' ')
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in words_to_remove]
    return ' '.join(filtered_words)

# Apply cleaning function to dataframes
tqdm.pandas()
overall_reviews_df['content'] = overall_reviews_df['content'].progress_apply(clean_text_combined)
customer_reviews_df['content'] = customer_reviews_df['content'].progress_apply(clean_text_combined)
```

#### 3.3.3: Removing Stopwords
- Download and load English stopwords from NLTK.
- Define a function `remove_stopwords` to filter out stopwords from the text.
- Apply this function to the 'content' column of both `overall_reviews_df` and `customer_reviews_df`.

```python
# Step 3: Removing Stopwords
# Download and load stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Split text into words and filter out stopwords
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Apply function to remove stopwords from content
tqdm.pandas()
overall_reviews_df['content'] = overall_reviews_df['content'].progress_apply(remove_stopwords)
customer_reviews_df['content'] = customer_reviews_df['content'].progress_apply(remove_stopwords)
```

#### 3.3.4: Lemmatization
    - Download necessary NLTK resources for lemmatization and tokenization.
    - Initialize the WordNetLemmatizer.
    - Define `lemmatize_row` to tokenize, POS tag, and lemmatize each word in a given row.
    - Use ThreadPoolExecutor to apply lemmatization in parallel for faster processing.
Note: if there is an error when running, rerun it, it will work afterwards.

```python
# Step 4: Lemmatization
# Function to map NLTK's POS tags to the format wordnet lemmatizer would understand
def get_wordnet_pos(word):
    """Map NLTK's POS tags to the format wordnet lemmatizer would understand"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
# Function to lemmatize each row
def lemmatize_row(row_tuple):
    # Extract text from the row
    _, row = row_tuple
    text = row['content']

    # Tokenize, POS tag, and lemmatize words
    words = nltk.word_tokenize(text)
    pos_tagged = [(word, get_wordnet_pos(word)) for word in words]
    lemmatized = [lemmatizer.lemmatize(word, pos) for word, pos in pos_tagged]

    # Rejoin the lemmatized text
    lemmatized_text = ' '.join(lemmatized)

    # Correct spacing around apostrophes for contractions (e.g., "wouldn't" -> "wouldn't")
    corrected_text = re.sub(r"\s’\s", "’", lemmatized_text)

    # Update the row content
    row['content'] = corrected_text
    return row

# Function to apply lemmatization in parallel
def apply_lemmatization_in_parallel(dataframe):
    with ThreadPoolExecutor(max_workers=16) as executor:
        results = list(tqdm(executor.map(lemmatize_row, dataframe.iterrows()), total=len(dataframe)))
    return pd.DataFrame(results)

# Apply lemmatization in parallel
overall_reviews_df = apply_lemmatization_in_parallel(overall_reviews_df)
customer_reviews_df = apply_lemmatization_in_parallel(customer_reviews_df)
```

## 4.1: Review-Based Features
### 4.1.1: Sentiment Analysis
- Conduct sentiment analysis on customer reviews to categorize sentiments as positive, neutral, or negative.
- Use NLP techniques to determine the overall sentiment expressed in each review.

In the sentiment analysis of customer reviews, I have chosen to use NLTK's SentimentIntensityAnalyzer with the VADER lexicon. This decision was influenced by VADER's ability to effectively interpret various linguistic expressions commonly found in customer reviews, including internet slang and emoticons. The analysis was performed on the complete, unfiltered text to retain the context and subtle nuances, essential for accurate sentiment interpretation. Filtering elements like stop words could potentially lead to losing critical sentiment-related information. VADER offers a 'compound' sentiment score, which ranges from -1 for highly negative to +1 for highly positive sentiment. It also provides proportions of negative, neutral, and positive sentiments. This approach helps categorize the overall sentiment of each review in a comprehensive manner, ensuring a nuanced understanding of the customers' sentiments toward the products.

Step 1: Create a Function for Sentiment Analysis

```python
def analyze_sentiment(text):
    # Initialize VADER
    sia = SentimentIntensityAnalyzer()

    # Analyze the sentiment of the text
    sentiment_scores = sia.polarity_scores(text)

    # Determine the overall sentiment
    if sentiment_scores['compound'] >= 0.05:
        return 'positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

Step 2: Apply Sentiment Analysis to Customer Reviews

```python
# Apply the analyze_sentiment function to the review content
tqdm.pandas()  # For progress bar
customer_reviews_df['sentiment'] = customer_reviews_df['content'].progress_apply(analyze_sentiment)
```

In our approach, we factored in the number of reviews each pen received as a weight, under the premise that a higher number of reviews could indicate a more reliable rating.

This figure reflects not just the raw average price, but also the popularity and customer satisfaction based on the review counts.

### 4.4.2: Average Specifications of Fountain Pens
- Compute the average specifications of fountain pens in terms of dimensions, weight, and other measurable attributes.

```python
# First, we will create a list of columns that contain measurable specifications
spec_columns = [
    'diameter_body', 'diameter_cap_without_clip', 'diameter_cap_with_clip',
    'diameter_grip_mm', 'length_body', 'length_cap', 'length_nib',
    'length_overall_closed', 'length_overall_posted', 'weight_body', 'weight_cap',
    'weight_overall_g', 'cap_rotations', 'max_ink_capacity'
]

# Some values may have measurement units in the strings, we need to remove these and convert to numeric.
# We will assume all measurements are given in millimeters or grams, and thus remove these units.
for col in spec_columns:
    # Convert the column to string
    product_details_df[col] = product_details_df[col].astype(str)

    # Extract the numeric part and convert to float
    product_details_df[col] = product_details_df[col].str.extract(r'(\d+\.?\d*)').astype(float)
```

##### Findings from 4.4.2: Average Specifications Analysis
The average specifications for fountain pens reveal that the most common characteristics focus significantly on length dimensions. The average length overall when posted is the greatest at approximately 157 mm, indicative of the importance of comfort and size when the pen is in use. Body length and closed length also have notable averages, underscoring their impact on the pen's portability and storage. In contrast, weights and diameters show lower average values, suggesting a preference for slimmer and lighter pens among users. The cap rotations and max ink capacity have the lowest averages, suggesting these features might vary less across different models or might be less crucial to consumers. This insight into fountain pen design preferences can guide manufacturers in aligning their products with consumer expectations.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating', y='sentiment_score', data=merged_df, hue='sentiment', alpha=0.6)

# Define the title and the labels of the axes
plt.title('Relationship between Ratings and Sentiment Scores')
plt.xlabel('Customer Rating')
plt.ylabel('Sentiment Score')

# Show the plot
plt.show()
```

### 5.1.2: Aggregated Ratings and Sentiment by Product
An aggregated view of ratings and sentiments at the product level will be presented to highlight top-performing products.
- A minimum of 50 reviews is set to ensure that the product ratings reflect the general consensus of the customers.

```python
# Group by product ID and calculate mean rating and sentiment score
product_grouped = merged_df.groupby('id').agg({
    'rating': 'mean',
    'sentiment_score': 'mean',
    'content': 'size'  # This counts the number of reviews per product
}).rename(columns={'content': 'review_count'}).reset_index()

# Merge with product details to get the title of the pen
product_details_aggregated = product_grouped.merge(
    product_details_df[['id', 'title', 'vendor', 'price']],
    on='id',
    how='left'
)

# Filter out products with less than 50 reviews
products_filtered = product_details_aggregated[product_details_aggregated['review_count'] >= 50]

# Now, sort by average rating and sentiment score to find top-performing products
top_products = products_filtered.sort_values(
    by=['rating', 'sentiment_score', 'review_count'],
    ascending=[False, False, False]
)

# Display the top 10 products
top_10_products = top_products.head(10)
top_10_products = top_10_products.copy()
top_10_products
```

```python
# Combine 'vendor' and 'title' to a single column for plotting
top_10_products['vendor_title'] = top_10_products['vendor'] + ' - ' + top_10_products['title']

# Plotting the average rating of top products
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='vendor_title', data=top_10_products, palette='coolwarm')
plt.title('Average Rating of Top Products')
plt.xlabel('Average Rating')
plt.ylabel('Product Title')
plt.show()

# Plotting the average sentiment score of top products
plt.figure(figsize=(10, 6))
sns.barplot(x='sentiment_score', y='vendor_title', data=top_10_products, palette='coolwarm')
plt.title('Average Sentiment Score of Top Products')
plt.xlabel('Average Sentiment Score')
plt.ylabel('Product Title')
plt.show()
```

#### Findings from 5.1.2: Aggregated Ratings and Sentiment by Product
The analysis of fountain pen reviews reveals that the Pilot Metropolitan and TWSBI ECO series, particularly the special editions, are top-rated by customers. Limited edition pens like the Kaweco Sport - Iridescent Pearl also show high customer satisfaction. Positive sentiment scores closely track with high ratings, indicating that customers’ feelings about a product are consistent with the numerical ratings they leave. Special and visually appealing editions tend to resonate well with customers, suggesting a preference for unique and aesthetically pleasing design features in fountain pens.
